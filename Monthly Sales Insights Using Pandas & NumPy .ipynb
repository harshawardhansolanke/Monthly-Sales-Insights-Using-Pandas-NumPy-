```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 1. Creating a Sample Dataset

data = {"Month": ["Jan", "Feb", "Mar", "Apr", "May", "Jun"],
        "Sales": [12000, 15000, 17000, 13000, 18000, 22000],
        "Expenses": [5000, 7000, 9000, 4000, 8000, 10000]
       }

df = pd.DataFrame(data)

print("----- Sales Data -----")
print(df)
print()
# 2. Basic Analysis Using Pandas

total_sales = df["Sales"].sum()
avg_sales = df["Sales"].mean()
max_sales = df["Sales"].max()

print("\nTotal Sales:", total_sales)
print("Average Sales:", avg_sales)
print("Highest Sales:", max_sales)

# 3. NumPy Calculations

profit = np.array(df["Sales"]) - np.array(df["Expenses"])
df["Profit"] = profit

print("\nProfit added using NumPy:")
print()
print(df)
print()

# 4. Visualization Using Matplotlib

plt.figure(figsize=(8,5))
plt.plot(df["Month"], df["Sales"], marker="o", label="Sales")
plt.plot(df["Month"], df["Expenses"], marker="o", label="Expenses")
plt.plot(df["Month"], df["Profit"], marker="o", label="Profit")

plt.title("Monthly Sales, Expenses & Profit")
plt.xlabel("Month")
plt.ylabel("Amount in ")
plt.legend()
plt.grid(True)
plt.show()
```
